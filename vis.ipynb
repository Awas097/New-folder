```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = sns.load_dataset('iris')
df
```

```python
# Bar plot for sepal length
_ = sns.barplot(x = 'species', y='sepal_length', data=df)
```

```python
_ = sns.barplot(x = 'species', y='sepal_width', data=df)
```

```python
_ = sns.barplot(x = 'species', y='petal_length', data=df)
```

```python
_ = sns.barplot(df, x = 'species', y='petal_width',)
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize=(10,10), nrows=2, ncols=2)
_ = sns.barplot(x = 'species', y='sepal_length', data=df, ax=ax[0,0])
_ = sns.barplot(x = 'species', y='sepal_width', data=df, ax=ax[0,1])
_ = sns.barplot(x = 'species', y='petal_length', data=df, ax=ax[1,0])
_ = sns.barplot(df, x = 'species', y='petal_width', ax=ax[1,1])
```

```python
fig,ax = plt.subplots(figsize=(40,10), ncols=4)
_ = sns.barplot(x = 'species', y='sepal_length', data=df, ax=ax[0])
_ = sns.barplot(x = 'species', y='sepal_width', data=df, ax=ax[1])
_ = sns.barplot(x = 'species', y='petal_length', data=df, ax=ax[2])
_ = sns.barplot(df, x = 'species', y='petal_width', ax=ax[3])
```
